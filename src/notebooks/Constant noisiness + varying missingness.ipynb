```python
import pandas as pd
import numpy as np
import random
```

```python
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
```

```python
noisy_df = pd.read_csv('../../data/train_set_noisiness_0.2.csv')
noisy_df.head()
```

```python
noisy_df.isna().sum()
```

```python
def insert_nans(df, col, frac):
    non_null_indices = df[df[col].notna()].index.tolist()
    to_change_indices = random.sample(non_null_indices, int(frac*len(non_null_indices)))
    df.loc[to_change_indices, col] = np.nan
    return df
```

```python
frac = 0.1
```

```python
for column in noisy_df.columns:
    if column not in ['hemoglobin', 'gender', 'label']:
        noisy_df = insert_nans(noisy_df, column, frac)
noisy_df.head()
```

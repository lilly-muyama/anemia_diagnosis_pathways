```python
import pandas as pd
import numpy as np
import random
```

```python
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
```

```python
def insert_nans(df, col, frac):
    '''
    Inserts missing values in a dataframe column
    '''
    non_null_indices = df[df[col].notna()].index.tolist() #getting not null indices
    to_change_indices = random.sample(non_null_indices, int(frac*len(non_null_indices)))
    df.loc[to_change_indices, col] = np.nan
    return df    
```

```python
frac = 0.1
```

```python
training_df = pd.read_csv('../../data/train_set_basic.csv')
training_df.head()
```

```python
for column in training_df.columns:
    if column not in ['hemoglobin', 'gender', 'label']:
        training_df = insert_nans(training_df, column, frac)
training_df.isna().sum()
```

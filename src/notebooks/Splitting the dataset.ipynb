```python
import pandas as pd
from sklearn.model_selection import train_test_split
import sys
sys.path.append('..')
from modules import constants
```

```python
SEED = 42
```

```python
all_df = pd.read_csv('../../data/full_anemia_dataset.csv')
all_df.head()
```

```python
class_dict = constants.CLASS_DICT
all_df['label'] = all_df['label'].replace(class_dict)
```

```python
X = all_df.iloc[:, 0:-1]
y = all_df.iloc[:, -1]
```

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=SEED)
X_train_val.shape, y_train_val.shape, X_test.shape, y_test.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.10, random_state=SEED, 
                                                  stratify=y_train_val)
X_train.shape, y_train.shape, X_val.shape, y_val.shape
```

```python
train_df = pd.concat([X_train, y_train], axis=1).reset_index(drop=True)
train_df = train_df.fillna(-1)
val_df = pd.concat([X_val, y_val], axis=1).reset_index(drop=True)
val_df = val_df.fillna(-1)
test_df = pd.concat([X_test, y_test], axis=1).reset_index(drop=True)
test_df = test_df.fillna(-1)
```
